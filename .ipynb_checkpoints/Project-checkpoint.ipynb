We can found column with `Standart Diviations` that equls to 0 and remove them becouse they cant imporve our model

```python
std_zero = data_cleaned.copy()
std_zero_std = std_zero.std()
std_zero_index = std_zero_std[std_zero_std == 0].index
std_zero = std_zero.drop(std_zero_index, axis=1)
std_zero
```

Delete Columns more than 70% of nan

```python
nan_percentages_columns = (std_zero.isnull().sum() / len(std_zero)) * 100
nan_percentages_rows = (std_zero.isnull().sum(axis=1) / len(std_zero.columns)) * 100
rows_to_delete = nan_percentages_rows[nan_percentages_rows > 70].index
columns_to_delete = nan_percentages_columns[nan_percentages_columns > 70].index
std_zero = std_zero.drop(columns_to_delete, axis=1)
std_zero = std_zero.drop(rows_to_delete, axis=0)
std_zero
```

Delete Less then 5%

```python
rows_to_delete = nan_percentages_rows[nan_percentages_rows < 5].index
std_zero = std_zero.drop(rows_to_delete, axis=0)
std_zero
```

For NaN Values pecentegae from 5\% to 15\% we will be input number form mean using sicikt learn

```python
missing_percentages = (std_zero.isnull().sum() / len(std_zero)) * 100
columns_to_impute  = missing_percentages[(missing_percentages >= 5) & (missing_percentages <= 15)].index
selected_columns = std_zero[columns_to_impute].copy()
imputer = SimpleImputer(strategy='mean')
selected_columns_imputed = pd.DataFrame(imputer.fit_transform(selected_columns), columns=selected_columns.columns)
selected_columns_imputed
std_zero[columns_to_impute] = selected_columns_imputed
std_zero
```

```python
columns_to_impute = missing_percentages[(missing_percentages >= 15) & (missing_percentages <= 70)].index
selected_columns = std_zero[columns_to_impute].copy()
imputer = IterativeImputer()
imputed_data = imputer.fit_transform(std_zero)
selected_columns_imputed = pd.DataFrame(imputer.fit_transform(selected_columns), columns=selected_columns.columns)
std_zero[columns_to_impute] = selected_columns_imputed
std_zero
```
